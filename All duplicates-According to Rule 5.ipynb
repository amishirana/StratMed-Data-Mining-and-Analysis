## Converted all entries to upper case

```python
x_upper=x.apply(lambda x: x.astype(str).str.upper())
x_upper
```

## Replaced NULL  values by NAN

```python
x_upper.replace("NULL","NAN",inplace=True)
```

```python
y = x_upper[['generic_name','form_name','dose_size','manufacturer_name','brand_name','unit_per_pack']]
y
```

##  Duplicates

```python
dup5 = y[y.duplicated(['generic_name','form_name','dose_size','manufacturer_name','brand_name','unit_per_pack'])]
dup5
```

```python
cols = ['generic_name','form_name','dose_size','manufacturer_name','brand_name','unit_per_pack']
```

## Created a new column combined which contains concatenation of 6 columns

```python
x_upper['combined'] = x_upper[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
```

## Dropped duplicate items from Combined column

```python
dm = x_upper.drop_duplicates('combined', keep=False)
dm
```

## List of index of Unique items

```python
l1 = list(dm.index)
l1
```

## Dropped index of unique items to get duplicates

```python
x_upper_dup = x_upper.drop(l1, axis=0)
x_upper_dup
```

```python
x_upper_dup= x_upper_dup.drop(['Unnamed: 0'], axis=1)
x_upper_dup
```
