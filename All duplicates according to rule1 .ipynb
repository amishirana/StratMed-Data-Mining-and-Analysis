```python
import numpy as np
import pandas as pd
```

```python
x = pd.read_csv("stratmed_item_master23-05.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
x
```

```python
y = x[['generic_name','form_name','dose_size','manufacturer_name','brand_name','unit_per_pack','mrp']]
y
```

##  Duplicates

```python
dup1 = y[y.duplicated()]
dup1
```

## Converted all entries to upper case

```python
dup1.apply(lambda x: x.astype(str).str.upper())
```

```python
x_upper=x.apply(lambda x: x.astype(str).str.upper())
x_upper
```

## Replaced NULL  values by NAN

```python
x_upper.replace('NULL','NAN',inplace=True)
```

```python
cols = ['generic_name','form_name','dose_size','manufacturer_name','brand_name','unit_per_pack','mrp']
```

## Created a new column combined which contains concatenation of 7 columns

```python
x_upper['combined'] = x_upper[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
```

```python
x_upper
```

```python
x_upper['combined']
```


## Dropped duplicate items from Combined column




```python
dm = x_upper.drop_duplicates('combined', keep=False)
dm
```

## List of index of Unique items

```python
l1 = list(dm.index)
l1
```

## Dropped index of unique items to get duplicates

```python
x_upper = x_upper.drop(l1, axis=0)
x_upper
```

```python
x_upper = x_upper.drop(['Unnamed: 0'], axis=1)
x_upper
```
